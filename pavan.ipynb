```python

    
def add(x,y):
    print(x+y)
def sub(x,y):
    print(x-y)
def mul(x,y):
    print(x*y)
def div(x,y):
    print(x/y)
x=int(input("enter first operand :"))
y=int(input("enter second operand :"))
z=input("enter the operand")
if z=="+":
    add(x,y)
elif z=="-":
    sub(x,y)
elif z=="*":
    mul(x,y)
elif z=="/":
    div(x,y)
```
